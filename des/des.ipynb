```python
def split(arr):
    return arr[0:(len(arr)//2)], arr[len(arr)//2:]

def initial_permutation(arr):
    return bitarray([arr[x - 1] for x in ip])

def rev_permutation(arr):
    return bitarray([arr[x - 1] for x in rev_ip])

def perm_1(arr):
    l = bitarray([arr[x - 1] for x in permuted_choice])
    return l

def get_keys(arr):
    ans = [split(arr)]
    for i in number_of_left_shifts:
        ans.append([ans[-1][0][i:] + ans[-1][0][:i], ans[-1][1][i:] + ans[-1][1][:i]])
    ans2 = []
    for i in ans[1:]:
        t = i[0] + i[1]
        ans2.append(bitarray([t[x - 1] for x in permuted_choice_2]))
    return ans2

def expand(arr):
    return bitarray([arr[x - 1] for x in expansion])

def encrypt_block(arr, keys):
    arr = initial_permutation(arr)
    l, r = split(arr)
    
    for i in keys:
        t = l
        l = r
        z = (expand(r)^i)
        r = bitarray()
        for j in range(len(z)//6):
            k = z[j*6:6*(j + 1)]
            a = 2*k[0] + k[-1]
            b = int(k[1:-1].to01(),2)
            r += bitarray("{0:{fill}4b}".format(s[j][a][b], fill='0'))
        r = bitarray([r[x - 1] for x in permutation])
        r = t^r
    t = r + l
    ans = rev_permutation(t)
    return ans

def decrypt_block(arr, keys):
    return encrypt_block(arr, reversed(keys))

def encrypt(m, key):
    bit_key = bitarray()
    bit_key.fromstring(key)
    if(len(bit_key) != 64): 
        print("Incorrect key")
        return
    keys = get_keys(perm_1(bit_key))

    arr = bitarray()
    arr.fromstring(m)
    while(len(arr) % 64 != 0):
        arr.append(False)
    c = bitarray()
    for i in range(0, len(arr), 64):
        c += encrypt_block(arr[i: i + 64], keys)
    return(c)

def decrypt(arr, key):
    bit_key = bitarray()
    bit_key.fromstring(key)
    if(len(bit_key) != 64): 
        print("Incorrect key")
        return
    keys = get_keys(perm_1(bit_key))
    c = bitarray()
    for i in range(0, len(arr), 64):
        c += decrypt_block(arr[i: i + 64], keys)
    while(c[-1:-9:-1] == bitarray("00000000")):
        c = c[:-8]
    return c.tostring()
    
```
